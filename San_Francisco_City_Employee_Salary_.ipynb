Data Cleaning

```python
salaries_df.info()
```

```python
salaries_df.isnull().sum()
```

```python
salaries_df.drop(['Status', 'Id', 'Notes', 'Agency'], axis=1, inplace=True)
```

```python
salaries_df.head(1)
```

```python
salaries_df.info()
```

```python
# convert numeric columns to float instead of object.
salaries_df['BasePay'] = pd.to_numeric(salaries_df['BasePay'],errors='coerce')
salaries_df['OvertimePay'] = pd.to_numeric(salaries_df['OvertimePay'],errors='coerce')
salaries_df['OtherPay'] = pd.to_numeric(salaries_df['OtherPay'],errors='coerce')
salaries_df['Benefits'] = pd.to_numeric(salaries_df['Benefits'],errors='coerce')
salaries_df['TotalPay'] = pd.to_numeric(salaries_df['TotalPay'],errors='coerce')
salaries_df['TotalPayBenefits'] = pd.to_numeric(salaries_df['TotalPayBenefits'],errors='coerce')

# convert text column to string
salaries_df['EmployeeName'] = salaries_df['EmployeeName'].astype('string')
salaries_df['JobTitle'] = salaries_df['JobTitle'].astype('string')

# confirm changes
salaries_df.info()
```

```python
# Mengubah tipe data kolom 'Year' menjadi datetime
salaries_df['Year'] = pd.to_datetime(salaries_df['Year'], format='%Y')


# confirm changes
salaries_df.info()
```

```python
# missing values

salaries_df.isnull().sum()
```

```python
salaries_df['Benefits'] = salaries_df['Benefits'].fillna(0.00)
salaries_df['OvertimePay'] = salaries_df['OvertimePay'].fillna(0.00)
salaries_df['OtherPay'] = salaries_df['OtherPay'].fillna(0.00)
```

```python
salaries_df[salaries_df.BasePay == 0].BasePay.count()
```

```python
salaries_df = salaries_df.drop(salaries_df[salaries_df.BasePay == 0].index)
```
